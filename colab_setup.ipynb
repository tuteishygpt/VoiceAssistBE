```python
# 2) Клануем або абнаўляем рэпазіторый
import os, subprocess
REPO_URL = "https://github.com/<your-org>/VoiceAssistBE.git"  # ← абнаві на свой URL
TARGET_DIR = "VoiceAssistBE"
if not os.path.exists(TARGET_DIR):
    subprocess.run(["git", "clone", REPO_URL, TARGET_DIR], check=True)
else:
    subprocess.run(["git", "-C", TARGET_DIR, "pull"], check=True)
os.chdir(TARGET_DIR)
print('Працоўная дырэкторыя:', os.getcwd())
```

```python
# 3) Ключы і параметры
import os
os.environ['GEMINI_API_KEY'] = 'PASTE_GEMINI_KEY_HERE'  # ← устаў свой ключ
os.environ['GEMINI_MODEL'] = os.environ.get('GEMINI_MODEL', 'gemini-1.5-flash')
NGROK_AUTH_TOKEN = ''  # неабавязкова, але пажадана для стабільнага тунэля
if NGROK_AUTH_TOKEN:
    from pyngrok import conf
    conf.get_default().auth_token = NGROK_AUTH_TOKEN
print('GEMINI_MODEL:', os.environ['GEMINI_MODEL'])
```

```python
# 4) Адкрываем тунэль на порт 8000
from pyngrok import ngrok
public_tunnel = ngrok.connect(8000, 'http')
print('Публічны URL для кліента:', public_tunnel.public_url)
```

```python
# 5) Запускаем FastAPI-сервер
import nest_asyncio
nest_asyncio.apply()
import uvicorn
from server import app
uvicorn.run(app, host='0.0.0.0', port=8000, log_level='info')
```
